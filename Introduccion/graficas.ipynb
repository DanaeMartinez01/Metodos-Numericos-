# Gr치fica de $f(x)=x - 3^{-x}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Funci칩n cuadr치tica.
def f(x):
    return x - 3**(-x)
```

```python
# Valores del eje X que toma el gr치fico.
x=np.linspace(-50, 50,50)
```

```python
#Graficar ambas funciones.
plt.plot(x, f(x))
plt.axhline(1, color="blue")
plt.axvline(0, color="black")
plt.xlim(-10, 10)
plt.ylim(-5, 5)
plt.savefig("output.png") # Guardar gr치fico como im치gen PNG.
# Mostrarlo.
plt.show()
```
